## Target 

The target corresponds to a loan that has been refund by the client (target value to 0) or not (target value to 1). Target variable (1 - client with payment difficulties: he/she had late payment more than X days on at least one of the first Y installments of the loan in our sample, 0 - all other cases)

```python
application_train["TARGET"].unique()
```

There is only two values for the TARGET column, as expected.

```python
sns.countplot(application_train["TARGET"])
plt.show()
```

```python
application_train["TARGET"].value_counts()
```

The target column is clearly imbalanced, with a large majority (282 686) of loans that are refund while a minority (24 825) is not refund. Training data will have to balanced, for example using the SMOTE library, during modelling.
